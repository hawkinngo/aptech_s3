```python
df = pd.read_csv('https://raw.githubusercontent.com/panicpotatoe/dataset/master/SOCR_Data_Dinov_020108_HeightsWeights.csv')
df.sample(3)
```

$$
r_{xy} = \frac{\sum_{i=1}^n (x_i - \overline{x})(y_i - \overline{y})}{\sqrt{\sum_{i=1}^n(x_i - \overline{x})^2 \sum_{i=1}^n(y_i - \overline{y})^2}} 
$$

```python
mean_weight = np.mean(df['Weight(Pounds)'])
mean_height = np.mean(df['Height(Inches)'])

def numerator(list_x, list_y, mean_x, mean_y):
    len_list = len(list_x)
    numerator = [(list_x[i] - mean_x) * (list_y[i] - mean_y) for i in range(len_list)]
    return np.sum(numerator)

def denominator(list_x, list_y, mean_x, mean_y):
    len_list = len(list_x)
    denominator_x = [(list_x[i] - mean_x)**2 for i in range(len_list)]
    denominator_y = [(list_y[i] - mean_y)**2 for i in range(len_list)]
    return np.sqrt(np.sum(denominator_x)*np.sum(denominator_y))

np_r = np.corrcoef(df['Weight(Pounds)'], df['Height(Inches)'])

print(f"{np_r=}")

numer = numerator(df['Height(Inches)'], df['Weight(Pounds)'], mean_height, mean_weight)
denom = denominator(df['Height(Inches)'], df['Weight(Pounds)'], mean_height, mean_weight)

print(numer/denom)
```

```python
df_500_data = pd.read_csv('https://raw.githubusercontent.com/panicpotatoe/dataset-practical-stats/main/sp500_data.csv')
```

```python
df_500_data.set_index("Unnamed: 0", inplace=True)
```

```python
df_500_data
```

```python
df500 = df_500_data

column_idx = {}
idx = 0
max_r = 0
com_1 = com_2 = ""

for column in df500:
    column_idx[idx] = column
    idx += 1

for col_index in range(len(df500.columns) - 1):
    col_name_curr = df500[column_idx[col_index]]
    col_name_next = df500[column_idx[col_index + 1]]

    mean_col_curr = np.mean(col_name_curr)
    mean_col_next = np.mean(col_name_next)


    numer = numerator(col_name_curr, col_name_next, mean_col_curr, mean_col_next)
    denom = denominator(col_name_curr, col_name_next, mean_col_curr, mean_col_next)
    
    r = numer / denom
    if r > max_r:
        com_1 = column_idx[col_index]
        com_2 = column_idx[col_index + 1]
        max_r = r

print(f"{com_1=}")
print(f"{com_2=}")
```

```python
max_r = 0
com_1 = com_2 = ""

for col_index in range(len(df500.columns) - 1):
    col_name_curr = df500[column_idx[col_index]]
    col_name_next = df500[column_idx[col_index + 1]]
    
    r = np.corrcoef(col_name_curr, col_name_next)[1, 0]

    if r > max_r:
        com_1 = column_idx[col_index]
        com_2 = column_idx[col_index + 1]
        max_r = r

print(f"{com_1=}")
print(f"{com_2=}")
```
