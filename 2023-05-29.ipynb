```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/panicpotatoe/dataset-practical-stats/main/f500.csv')
df.sample(5)
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
industry_list = df["industry"].unique()
print(f"{industry_list=}")
print(f"{len(industry_list)}")
```

```python
sector_list = df['sector'].unique()
print(f'{sector_list=}')
print(f'{len(sector_list)}')
```

```python
pivot_industry_company = df[['industry', 'sector', 'company']].pivot_table(index='industry', values='sector', aggfunc='count', fill_value=0)
pivot_industry_company
```

```python
df.info()
```

```python
any(df.duplicated())
```

```python
pivot_industry_country = df.pivot_table(index='industry', values='country', aggfunc='count').sort_values('country', ascending=False)
pivot_industry_country
```

```python
pivot_company_profit_desc = df.pivot_table(index='company', values='profits', aggfunc='sum').sort_values('profits', ascending=False).head(10)
pivot_company_profit_desc
```

```python
pivot_company_profit_desc_USA = df[df['country'] == 'USA'].pivot_table(index='company', values='profits', aggfunc='sum').sort_values('profits', ascending=False).head(10)
pivot_company_profit_desc_USA
```

```python
pivot_company_insection = pd.merge(pivot_company_profit_desc, pivot_company_profit_desc_USA)
pivot_company_insection
```

```python
df['change_rank'] = df['rank'] - df['previous_rank']
df
```

```python
pd.set_option('display.max_rows', False)
df_jump_rank = df.sort_values(
    by=['change_rank', 'rank'],
    ascending=[True, True]
)
df_jump_rank[['rank', 'previous_rank', 'change_rank']]
```

```python
df_jump_rank.head(5)
```

```python
df_japan_top_emp = df[df['hq_location'].str.contains('Japan')].sort_values('employees', ascending=False)
df_japan_top_emp.head(5)
```

```python
df_max_emp_by_country = df.sort_values("employees", ascending=False).drop_duplicates(
    subset='country',
    keep="first"
)
df_max_emp_by_country[['country', 'company', 'employees']]
```

```python
pd.set_option('display.max_rows', False)
df_max_emp_by_country = df.sort_values(
    by=["profits", "employees"], ascending=[False, True]
).drop_duplicates(subset="country", keep="first")
df_max_emp_by_country[['country', 'company', 'employees']].head(5)
```

```python
df_max_profit_change_emp_by_country = df.sort_values(
    by=["profit_change", "revenue_change", "employees"], ascending=[False, False, True]
).drop_duplicates(subset="country", keep="first")
df_max_profit_change_emp_by_country
```
