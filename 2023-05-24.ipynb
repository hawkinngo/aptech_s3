```python
df_group_by_year = df.groupby('work_year')
```

```python
df_2023_group = df_group_by_year.get_group(2023)
df_2023_group
```

```python
df_drop_dups = df.drop_duplicates(keep='first')
df_drop_dups
```

```python
df_2023_drop = df_drop_dups.loc[df['work_year'] == 2023]
df_2023_drop
```

```python
df_old_drop = df_2023_drop.merge(df_drop_dups, how="outer")
df_old_drop
```

```python
job_title = df['job_title'].unique()
job_title_to_num = [(job_title[i], i) for i in range(len(job_title)) ]
job_title_to_num
```

```python
df_job_title = pd.DataFrame(data=job_title_to_num, columns=['job_title', 'job_encode'])
df_job_title
```

```python
df_endcode = df.merge(df_job_title, how="left")
df_endcode
```
