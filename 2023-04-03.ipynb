```python
# 1000 random numbers in normal distribution between 0 and 10
list1 = np.random.normal(loc=5, scale=2, size=1000)
list1

plt.hist(list1, bins=10)
plt.show()
```

```python
min_list1 = np.min(list1)
max_list1 = np.max(list1)
mean_list1 = np.mean(list1)
median_list1 = np.median(list1)
print(f"{min_list1=},{max_list1=},{mean_list1=},{median_list1=}")
```

```python
list2 = np.random.normal(loc=0, scale=0.1, size=1000)
outliers = np.array([4, 5, 4.5, 6])
list2 = np.append(list2, outliers)

plt.hist(list2, bins=10)
plt.show()
```

```python
median_list2 = np.median(list2)
median_list2

diff = np.abs(list2 - median_list2)
mean_diff = np.mean(diff)
print(f"{mean_diff=}")

outlier_list = list2[diff > 5 * mean_diff]
print(f"{outlier_list=}")
```

$$\text{mean absolute deviations}=\frac{\sum^n_{i=1}|x_i - \bar{x}|}{n}$$


```python
# Create a function for mean absolute deviations with in is a list
def mean(x):
    sum_x = 0
    [sum_x := sum_x + i for i in x]

    return sum_x / len(x)


def mean_absolute_deviation(x):
    mean_x = mean(x)
    diff = np.abs(x - mean_x)
    mean_diff = np.mean(diff)
    return mean_diff


mean_absolute_deviation(list1)
```

```python
import pandas as pd

df = pd.read_csv(
    "https://raw.githubusercontent.com/panicpotatoe/dataset-practical-stats/main/state.csv"
)
df.sample(5)
```

```python
# Calculte the mean of the population using numpy
mean_population = np.mean(df["Population"])
print(f"{mean_population=}")
```

```python
# mean_abs_dev_population =
devi = 0
[devi := devi + np.abs(mean_population - pop) for pop in df["Population"]]
mean_abs_devi = devi / len(df["Population"])
print(f"{mean_abs_devi=}")
```

```python
mean_abs_dev = mean_absolute_deviation(df["Population"])
print(f"{mean_abs_dev=}")
```

$$\text{mean absolute deviations}=\frac{\sum^n_{i=1}|x_i - \bar{x}|}{n}$$
$$\text{variance}=\frac{\sum^n_{i=1}(x_i - \bar{x})^2}{n-1}$$
$$\text{standard deviations}=\sqrt\frac{\sum^n_{i=1}(x_i - \bar{x})^2}{n-1}$$


```python
sqr_dev = 0
sqr_dev2 = 0
[sqr_dev := sqr_dev + np.square(i - mean_population) for i in df["Population"]]
[sqr_dev2 := sqr_dev2 + (i - mean_population)**2 for i in df["Population"]]

variance = sqr_dev / (len(df["Population"]) - 1)
std_dev = np.sqrt(variance)

print(f"{sqr_dev=}")
print(f"{sqr_dev2=}")
print(f"{variance=}")
print(f"{std_dev=}")
```
